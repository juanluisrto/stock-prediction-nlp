```python
def split_by_date(df, date):
    return df[:date].shift(-1).dropna(), df[date:]
```

```python
def lstm_model(n_steps):
    n_features = 5
    model = Sequential()
    model.add(LSTM(100, input_shape = (n_steps, n_features), return_sequences = True))
    model.add(TimeDistributed(Dense(20, activation='elu')))
    model.add(Flatten())
    model.add(Dense(1, activation='elu'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    print(model.summary())
    return model
```

```python
def dummy_series(length, dim, as_df = True):
    df = pd.DataFrame(np.array([dim*[i] for i in range(length)]),pd.date_range("2019-01-01", periods = length))
    return df if as_df else df.values
def generate_labels(array, window, index):
    return array[index][window:]
```
